```python
import os
```

```python
strain_list=os.listdir("./Leafly-scraper-master/results/")
```

```python
from pymongo import MongoClient
from pprint import pprint

client = MongoClient()

db = client.strain_project

db.strains
```

```python
finished_csv=[]
```

```python
error_csvs=[]
```

```python
# counter=0
for strain_path in strain_list:
    flag = False
    if strain_path in finished_csv:
        None
    else:

        strain_full_path="./Leafly-scraper-master/results/"+strain_path    
        #print (strain_full_path)
        try:
            df=pd.read_csv(strain_full_path, header=None,skiprows=2,usecols=[0,1,2])
            df.columns=['username','url','review']
            df['strain_name']=strain_path[:-4]
            row_dict_list=df.apply(lambda x: x.to_dict(),axis=1).values
            flag=True
        except:
            error_csvs.append(strain_path)
            continue
            
        if flag:
            db.strains.insert_many(row_dict_list.tolist())
            finished_csv.append(strain_path)
            print(strain_path + " complete")
        
        
print(len(error_csvs))
        
#     counter+=1
```
